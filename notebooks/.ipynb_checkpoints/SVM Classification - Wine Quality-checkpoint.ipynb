## SVM is normally a single class classifier
In order to use it as a multiclass classifier special care is needed

## Train Test Split - Single Class

```python
from sklearn.model_selection import train_test_split
```

```python
# Convert wine quality into good & bad
# Bad Quality (3-6) as 0
# Good Quality (7-8) as 1
wineQuality['mod_quality'] = (wineQuality['quality'] >=7)*1
wineQuality.head()
```

```python
X = wineQuality.drop(['quality', 'mod_quality'],axis=1)
y = wineQuality['mod_quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

## Train SVM Classifier - Single Class

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
# Basic Training
svc_model.fit(X_train,y_train)
```

## Predictions & Evaluations - Single Class

```python
predictions = svc_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

## Gridsearch - Single Class

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001]} 
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=2)
grid.fit(X_train,y_train)
```

Best Params

```python
grid.best_params_
```

Best Estimator

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```

## Applying SVM to MultiClass

## Train Test Split - Multi Class

```python
X = wineQuality.drop(['quality', 'mod_quality'],axis=1)
y = wineQuality['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

## Train SVM Classifier - Multi Class

```python
from sklearn.svm import SVC
```

```python
svc_model_multiclass = SVC()
```

```python
# MutliClass Training
# set decision_function_shape='ovo' ; 'ovo' --> one vs. one, this will create multiple classifiers to run against each other
svc_model_multiclass.fit(decision_function_shape='ovo', X_train, y_train)
```

## Predictions & Evaluations - Multi Class

```python
predictions = svc_model_multiclass.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

## Data Import

```python
features=pd.read_csv('../data/retail-data-analytics/Features data set.csv')
features.head()
```

```python
sales=pd.read_csv('../data/retail-data-analytics/sales data-set.csv')

sales = sales.drop(columns=['Dept']).groupby(by=['Store', 'Date', 'IsHoliday'], as_index=False).sum()

sales.head()
# sales[sales['Store']==1]
# sales['Dept'].value_counts()
# sales.sort_values('Date', ascending=True)
```

```python
stores=pd.read_csv('../data/retail-data-analytics/stores data-set.csv')
stores.head()
```

## Train - Test Split

```python
from sklearn.model_selection import train_test_split

X = retailData.drop(columns=['Weekly_Sales', 'Date'])

X = sm.add_constant(X)

y = retailData['Weekly_Sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# Merge Back Training Data to use in statsmodel
# statsTrain = X_train.merge(pd.DataFrame(y_train))
statsTrain = X_train.join(pd.DataFrame(y_train))
statsTrain.head()
```

## Train Model

```python
### STATSMODELS ###

# create a fitted model using all Xs
# lm = smf.ols(formula='Weekly_Sales ~ .', data=statsTrain).fit()
lm = sm.OLS(y_train, X_train).fit()

# print a summary of the fitted model
lm.summary()
```

## Model Evaluation

```python
predictions = lm.predict(X_test)

# calculate these metrics by hand!
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

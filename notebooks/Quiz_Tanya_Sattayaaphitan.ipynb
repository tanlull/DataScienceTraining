## <span style="color:blue">Cleansing data: Replace Avg_Session_Length with mean</span>

```python
customers['Avg_Session_Length'].mean()
```

```python
def impute_Avg_Session_Length(x):
    if pd.isnull(x):
        return 33.052949561754026
    else:
        return x
    
customers['Avg_Session_Length'] = customers['Avg_Session_Length'].apply(impute_Avg_Session_Length)


customers.isnull().any()
```

```python
customers['Avg_Session_Length'].mean()
```

## <span style="color:blue">Continue Cleansing data on the following columns </span>

* Replace Time_on_App
- Time_on_Website
- Length_of_Membership
- Yearly_Amount_Spent with mean

```python
customers['Time_on_App'].mean()
```

```python
def impute_Time_on_App(x):
    if pd.isnull(x):
        return 12.05267420326762
    else:
        return x
    
customers['Time_on_App'] = customers['Time_on_App'].apply(impute_Time_on_App)
customers.isnull().any()
```

```python
customers['Time_on_Website'].mean()
```

```python
def impute_Time_on_Website(x):
    if pd.isnull(x):
        return 37.06053902837023
    else:
        return x
    
customers['Time_on_Website'] = customers['Time_on_Website'].apply(impute_Time_on_Website)
customers.isnull().any()
```

```python
customers['Length_of_Membership'].mean()
```

```python
def impute_Length_of_Membership(x):
    if pd.isnull(x):
        return 3.533547619971888
    else:
        return x
    
customers['Length_of_Membership'] = customers['Length_of_Membership'].apply(impute_Length_of_Membership)
customers.isnull().any()
```

```python
customers['Yearly_Amount_Spent'].mean()
```

```python
def impute_Yearly_Amount_Spent(x):
    if pd.isnull(x):
        return 499.3143867294591
    else:
        return x
    
customers['Yearly_Amount_Spent'] = customers['Yearly_Amount_Spent'].apply(impute_Yearly_Amount_Spent)
customers.isnull().any()
```

## Training and Testing Data

Now that we've explored the data a bit, let's go ahead and split the data into training and testing sets.
** Set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column. **

```python
X = customers.drop(columns=['Yearly_Amount_Spent','Email','Address','Avatar'],axis=1)
```

```python
y = customers['Yearly_Amount_Spent']
```

```python
X.head()
```

```python
y.head()
```

** Use model_selection.train_test_split from sklearn to split the data into training and testing sets. Set test_size=0.3 and random_state=101**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=101)
custTrain = X_train.join(pd.DataFrame(y_train))
custTrain.head()
```

## Training the Model

Now its time to train our model on our training data!

** Import LinearRegression from sklearn.linear_model **

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import statsmodels.formula.api as smf
from scipy import stats
```

**Create an instance of a LinearRegression() model named lm.**

```python
lm = smf.ols(formula='Yearly_Amount_Spent ~ Avg_Session_Length+Time_on_App+Time_on_Website+Length_of_Membership', data=custTrain)
```

** Train/fit lm on the training data.**

```python
lm = lm.fit()
```

**Print out the coefficients of the model**

```python
# The coefficients
lm.summary()
```

## Predicting Test Data
Now that we have fit our model, let's evaluate its performance by predicting off the test values!

** Use lm.predict() to predict off the X_test set of the data.**

```python
predictions = lm.predict(X_test)
```

** Create a scatterplot of the real test values versus the predicted values. **

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

## Evaluating the Model

Let's evaluate our model performance by calculating the residual sum of squares and the explained variance score (R^2).

** Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error. Refer to the lecture or to Wikipedia for the formulas**

```python
# calculate these metrics by hand!
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

## Residuals

You should have gotten a very good model with a good fit. Let's quickly explore the residuals to make sure everything was okay with our data. 

**Plot a histogram of the residuals and make sure it looks normally distributed. Use either seaborn distplot, or just plt.hist().**

```python
sns.distplot((y_test-predictions),bins=50);
```
